```python
import numpy as np
from tensorflow import keras
img_path = keras.utils.get_file("cat.jpg", "https://img-datasets.s3.amazonaws.com/cat.jpg")

def get_img_array(img_path, size):
    img = keras.utils.load_img(img_path, target_size=size)
    array = keras.utils.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    return array

img_tensor = get_img_array(img_path, size = (128,128))
```

```python
activation = feature_extractor(
    keras.applications.xception.preprocess_input(img_tensor)
)
activation.shape
```

```python
def compute_loss(image, filter_index):
    activation = feature_extractor(image)
    filter_activation = activation[:, 2:-2, 2:-2, filter_index]
    return tf.reduce_mean(filter_activation)
```

```python
import tensorflow as tf
@tf.function
def gradient_ascent_step(image, filter_index, learning_rate):
    with tf.GradientTape() as tape:
        tape.watch(image)
        loss = compute_loss(image, filter_index)
    grads = tape.gradient(loss,image)
    grads = tf.math.l2_normalize(grads)
    image += learning_rate * grads
    return image
```

```python
img_width = 200
img_height = 200

def generate_filter_pattern(filter_index):
  iterations = 30
  learning_rate = 10.
  image = tf.random.uniform(
      minval = 0.4,
      maxval = 0.6,
      shape = (1, img_width, img_height, 3)
  )
  for i in range(iterations):
    image = gradient_ascent_step(image, filter_index, learning_rate)
  return image[0].numpy()
```

```python
def deprocess_image(image):
  image -= image.mean()
  image /= (image.std() + 1e-5) # Add epsilon to prevent division by zero
  image *= 64
  image += 128
  image = np.clip(image, 0, 255).astype("uint8")
  image = image[25:-25, 25:-25, :]
  return image
```

```python
import matplotlib.pyplot as plt
plt.imshow(deprocess_image(generate_filter_pattern(filter_index=2)))
```

```python
all_images = []
for filter_index in range(64):
  print(f"Processing filter {filter_index}")
  image = deprocess_image(
      generate_filter_pattern(filter_index)
  )
  all_images.append(image)
margin = 5
n = 8
cropped_width = img_width - 25*2
cropped_height = img_height - 25*2
width = n * cropped_width + (n-1) * margin
height = n * cropped_height + (n-1) * margin
stitched_filters = np.zeros((width, height, 3))

for i in range(n):
  for j in range(n):
    img = all_images[i * n + j]
    row_start = (cropped_width + margin) * i
    row_end = (cropped_width + margin) * i + cropped_width
    col_start = (cropped_height + margin) * j
    col_end = (cropped_height + margin) * j + cropped_height

    stitched_filters[row_start:row_end, col_start:col_end, :] = img
keras.utils.save_img(f"filter_layers.png", stitched_filters)
plt.imshow(stitched_filters.astype("uint8"))
plt.axis("off")
plt.show()
```
